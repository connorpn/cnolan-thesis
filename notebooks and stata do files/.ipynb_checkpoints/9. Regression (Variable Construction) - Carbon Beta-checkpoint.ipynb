# 7. Create PMC Factor 

```python
"IMPORT MODULES"
import os
import pandas as pd
import numpy as np

"SET THE WORKING DIRECTORY BELOW TO THE LOCATION OF DATA FILES"

working_directory = 'C:/Users/conno/OneDrive/University Study/Honours Thesis/cnolan-thesis' #set location using back slashes

os.chdir(working_directory)

print("Current working directory: {0}".format(os.getcwd()))


def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
            output_path = os.makedirs(directory)
            print(output_path)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
        

# Folder where outputs will be saved (by default a folder within the working directory) 
createFolder('./output/') 
output_path = working_directory +'./output/'

print('Set WD: Done')
```

```python
"Import Data"

pmc_log_scope1 = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_log_scope1.csv')
pmc_log_scope2 = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_log_scope2.csv')
pmc_log_total_emissions = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_log_total_emissions.csv')
pmc_log_energy_consumption = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_log_energy_consumption.csv')

pmc_change_scope1 = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_change_scope1.csv')
pmc_change_scope2 = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_change_scope2.csv')
pmc_change_total_emissions = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_change_total_emissions.csv')
pmc_change_energy_consumption =  pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_change_energy_consumption.csv')

pmc_scope1_int =  pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_scope1_int.csv')
pmc_scope2_int =  pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_scope2_int.csv')
pmc_total_emissions_int =  pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_total_emissions_int.csv')
pmc_energy_consumption_int =  pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/pmc_energy_consumption_int.csv')


cross_sectional_returns = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/output/carbon_emissions_stock_returns_vars.csv')
fama_french_5_factors = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/data/Asia_Pacific_ex_Japan_5_Factors.csv')
fama_french_mom_factor = pd.read_csv ('https://raw.githubusercontent.com/connorpn/cnolan-thesis/main/data/Asia_Pacific_ex_Japan_MOM_Factor.csv')
```

```python
"Filter fama-french factors by the sample period"
fama_french_5_factors = fama_french_5_factors[(fama_french_5_factors['year-month'] >= 200807) & (fama_french_5_factors['year-month'] <= 202106)]
fama_french_mom_factor = fama_french_mom_factor[(fama_french_mom_factor['year-month'] >= 200807) & (fama_french_mom_factor['year-month'] <= 202106)]
```

```python

"log_scope1 Merge Fama French Factors"

carbon_beta_log_scope1_vars = pd.merge(pmc_log_scope1[['year','month','pmc']], fama_french_5_factors[['year-month','year', 'month', 'MKT-RF', 'SMB', 'HML']], how='left', on=['year','month'])
carbon_beta_log_scope1_vars = pd.merge(carbon_beta_log_scope1_vars, fama_french_mom_factor[['year', 'month', 'WML']], how='left', on=['year','month'])
carbon_beta_log_scope1_vars = pd.merge(carbon_beta_log_scope1_vars, cross_sectional_returns[['ticker_year_month','ticker','year','month','ret']], how='right', on=['year','month'])

obs_before = len(carbon_beta_log_scope1_vars)

carbon_beta_log_scope1_vars = carbon_beta_log_scope1_vars.dropna(subset=['ret', 'MKT-RF', 'SMB', 'HML', 'WML', 'pmc']).reset_index(drop=True)

obs_after = len(carbon_beta_log_scope1_vars)
obs_dropped = (obs_before - obs_after)
print('dropna(variables): ' + str(obs_dropped))

carbon_beta_log_scope1_vars = carbon_beta_log_scope1_vars[['ticker_year_month','ticker', 'year-month', 'year','month', 'ret', 'MKT-RF', 'SMB', 'HML', 'WML', 'pmc']]


"Save carbon_beta_log_scope1_vars"
output_filename = 'carbon_beta_log_scope1_vars.csv'
outputname = output_path + output_filename
carbon_beta_log_scope1_vars.to_csv(outputname, mode='w', index=False)
print("Exported File: " + outputname)



"log_scope2 Merge Fama French Factors"

carbon_beta_log_scope2_vars = pd.merge(pmc_log_scope2[['year','month','pmc']], fama_french_5_factors[['year-month','year', 'month', 'MKT-RF', 'SMB', 'HML']], how='left', on=['year','month'])
carbon_beta_log_scope2_vars = pd.merge(carbon_beta_log_scope2_vars, fama_french_mom_factor[['year', 'month', 'WML']], how='left', on=['year','month'])
carbon_beta_log_scope2_vars = pd.merge(carbon_beta_log_scope2_vars, cross_sectional_returns[['ticker_year_month','ticker','year','month','ret']], how='right', on=['year','month'])

obs_before = len(carbon_beta_log_scope2_vars)

carbon_beta_log_scope2_vars = carbon_beta_log_scope2_vars.dropna(subset=['ret', 'MKT-RF', 'SMB', 'HML', 'WML', 'pmc']).reset_index(drop=True)

obs_after = len(carbon_beta_log_scope2_vars)
obs_dropped = (obs_before - obs_after)
print('dropna(variables): ' + str(obs_dropped))

carbon_beta_log_scope2_vars = carbon_beta_log_scope2_vars[['ticker_year_month','ticker', 'year-month', 'year','month', 'ret', 'MKT-RF', 'SMB', 'HML', 'WML', 'pmc']]


"Save carbon_beta_log_scope2_vars"
output_filename = 'carbon_beta_log_scope2_vars.csv'
outputname = output_path + output_filename
carbon_beta_log_scope2_vars.to_csv(outputname, mode='w', index=False)
print("Exported File: " + outputname)


"log_total_emissions Merge Fama French Factors"

carbon_beta_log_total_emissions_vars = pd.merge(pmc_log_total_emissions[['year','month','pmc']], fama_french_5_factors[['year-month','year', 'month', 'MKT-RF', 'SMB', 'HML']], how='left', on=['year','month'])
carbon_beta_log_total_emissions_vars = pd.merge(carbon_beta_log_total_emissions_vars, fama_french_mom_factor[['year', 'month', 'WML']], how='left', on=['year','month'])
carbon_beta_log_total_emissions_vars = pd.merge(carbon_beta_log_total_emissions_vars, cross_sectional_returns[['ticker_year_month','ticker','year','month','ret']], how='right', on=['year','month'])

obs_before = len(carbon_beta_log_total_emissions_vars)

carbon_beta_log_total_emissions_vars = carbon_beta_log_total_emissions_vars.dropna(subset=['ret', 'MKT-RF', 'SMB', 'HML', 'WML', 'pmc']).reset_index(drop=True)

obs_after = len(carbon_beta_log_total_emissions_vars)
obs_dropped = (obs_before - obs_after)
print('dropna(variables): ' + str(obs_dropped))

carbon_beta_log_total_emissions_vars = carbon_beta_log_total_emissions_vars[['ticker_year_month','ticker', 'year-month', 'year','month', 'ret', 'MKT-RF', 'SMB', 'HML', 'WML', 'pmc']]


"Save carbon_beta_log_total_emissions_vars"
output_filename = 'carbon_beta_log_total_emissions_vars.csv'
outputname = output_path + output_filename
carbon_beta_log_total_emissions_vars.to_csv(outputname, mode='w', index=False)
print("Exported File: " + outputname)


"log_energy_consumption Merge Fama French Factors"

carbon_beta_log_energy_consumption_vars = pd.merge(pmc_log_energy_consumption[['year','month','pmc']], fama_french_5_factors[['year-month','year', 'month', 'MKT-RF', 'SMB', 'HML']], how='left', on=['year','month'])
carbon_beta_log_energy_consumption_vars = pd.merge(carbon_beta_log_energy_consumption_vars, fama_french_mom_factor[['year', 'month', 'WML']], how='left', on=['year','month'])
carbon_beta_log_energy_consumption_vars = pd.merge(carbon_beta_log_energy_consumption_vars, cross_sectional_returns[['ticker_year_month','ticker','year','month','ret']], how='right', on=['year','month'])

obs_before = len(carbon_beta_log_energy_consumption_vars)

carbon_beta_log_energy_consumption_vars = carbon_beta_log_energy_consumption_vars.dropna(subset=['ret', 'MKT-RF', 'SMB', 'HML', 'WML', 'pmc']).reset_index(drop=True)

obs_after = len(carbon_beta_log_energy_consumption_vars)
obs_dropped = (obs_before - obs_after)
print('dropna(variables): ' + str(obs_dropped))

carbon_beta_log_energy_consumption_vars = carbon_beta_log_energy_consumption_vars[['ticker_year_month','ticker', 'year-month', 'year','month', 'ret', 'MKT-RF', 'SMB', 'HML', 'WML', 'pmc']]


"Save carbon_beta_log_energy_consumption_vars"
output_filename = 'carbon_beta_log_energy_consumption_vars.csv'
outputname = output_path + output_filename
carbon_beta_log_energy_consumption_vars.to_csv(outputname, mode='w', index=False)
print("Exported File: " + outputname)
```
